```python
import os
import cv2 as cv
```

```python
def resize(img, scale = 0.1):
    width = int(img.shape[1] * scale)
    height = int(img.shape[0] * scale)
    dimensions = (width, height)
    return cv.resize(img, dimensions, interpolation=cv.INTER_AREA)

def pi_sigma(mat1, mat2):
    pi_mat = []
    if len(mat1) == len(mat2):
        for i in range(len(mat1)):
            pi_mat.append(mat1[i]*mat2[i])
        sigma = sum(pi_mat)
        return (pi_mat, sigma)
    else:
        print("Error")
        response = f"Length of Matrix: {mat1} and Matrix: {mat2} are not equal."
        return response
```

```python
img_DIR = r"B:\Projects\Neural Networks\Dataset\Triangles\Test"
final_DIR = r"B:\Projects\Neural Networks\Dataset\Triangles\Test"
```

```python
img_list = os.listdir(img_DIR)
print(img_list)
print(len(img_list))
for p in range(len(img_DIR)):
    try:
        if p == len(img_list):
            break
        img = cv.imread(os.path.join(img_DIR, img_list[p]))
        img_resize = resize(img)
        bin_img = []
        for i in range (len(img_resize)):
            row_list = []
            for j in range(len(img_resize[i])):
                if img_resize[i][j][0] != 0:
                    row_list.append(1)
                else:
                    row_list.append(0)
            bin_img.append(row_list)
        file_name = img_list[p].split(".")
        file = open(os.path.join(final_DIR, f"{file_name[0]}.txt"), "w")
        for i in range(len(bin_img)):
            for j in range(len(bin_img[i])):
                file.write(str(bin_img[i][j]))
            """if i != len(bin_img) - 1:
                file.write("b")"""
        file.close()
    except:
        print(f"Not an image: {img_list[p]}")
        print("Skipped...")
```
