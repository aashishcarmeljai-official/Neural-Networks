```python
import cv2 as cv
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
root_DIR = r"B:\Projects\Neural Networks\Dataset\Circles"
```

```python
def imgshow(img):
    image = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.show()
```

```python
def resize(img, scale = 0.1):
    width = int(img.shape[1] * scale)
    height = int(img.shape[0] * scale)

    dimensions = (width, height)
    return cv.resize(img, dimensions, interpolation=cv.INTER_AREA)
```

```python
img = cv.imread(os.path.join(root_DIR, "1.png"))
imgshow(img)
```

```python
img_thresh = cv.threshold(img, 0, 255, cv.THRESH_BINARY_INV)
imgshow(img_thresh[1])
```

```python
files = os.listdir(root_DIR)
print(files)
for i in range(len(files)):
    try:
        img = cv.imread(os.path.join(root_DIR, files[i]))
        img_thresh = cv.threshold(img, 0, 255, cv.THRESH_BINARY_INV)
        cv.imwrite(f"{root_DIR}\Inverted\{files[i]}", img_thresh[1])
    except:
        print(f"Not an image: {files[i]}")
```
