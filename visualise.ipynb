```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
file = open(r"B:\Projects\Neural Networks\Dataset\Triangles\Inverted\Binary\1.txt", "r")
img_mat = list(map(int, file.read()))
img_readable_mat = [] #115
k = 0
row = []
for i in range(len(img_mat)):
    if k < 115:
        row.append(img_mat[i])
        k = k + 1
    else:
        row.append(img_mat[i])
        img_readable_mat.append(row)
        row = []
        k = 0
for r in img_readable_mat:
    print(r)
file.close()
```
